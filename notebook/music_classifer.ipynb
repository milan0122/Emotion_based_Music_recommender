```python
import pandas as pd

def  Music_classifier(pred_class, song_df):
    # Mapping emotions to moods
    EMP = {
        'happy': ['Happy','Energetic],
        'fear': ['Calm'],             
        'sad': ['Sad', 'Happy','Calm'],     
        'angry': ['Calm'],       
        'neutral': ['Calm','Happy','Sad','Energetic],
        'surprise': ['Energetic','happy'],   
        'disgust': ['Sad']            
    }
    
    # Get the corresponding moods for the predicted emotion
    moods = EMP.get(pred_class.lower())
    
    if moods is None:
        print("Emotion not recognized.")
        return None

    song_df = song_df[song_df['mood'].isin(moods)]
    
    # Sort songs by popularity and select top 5
    top_songs = song_df.sort_values('popularity', ascending=False).head(5)
    
    return top_songs

# Example usage
# top_songs_df = get_top_songs('sad', song_df)
# display(top_songs_df)
```
